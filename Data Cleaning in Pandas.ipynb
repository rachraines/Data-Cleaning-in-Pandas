```python
import pandas as pd
```

```python
# Load in excel worksheet data set
df = pd.read_excel(r'C:\Users\rfros\Downloads\Customer Call List.xlsx')
df
```

```python
# Remove duplicates
df = df.drop_duplicates()
df
```

```python
# Delete unused column
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# Standardize Last_Name column
df["Last_Name"] = df["Last_Name"].str.strip("/")
df["Last_Name"] = df["Last_Name"].str.strip("...")
df["Last_Name"] = df["Last_Name"].str.strip("_")
df
```

```python
# Standardize Phone_Number

# Removes all non-alpha numeric values
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)

# Convert all values to strings
df["Phone_Number"] = df["Phone_Number"].astype(str)

# Format into xxx-xxx-xxxx
df["Phone_Number"] = df["Phone_Number"].str.replace(
    r"(\d{3})(\d{3})(\d{4})",
    r"\1-\2-\3",
    regex=True
)

# Remove Na and NaN values
df["Phone_Number"] = df["Phone_Number"].str.replace('Na','',regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('NaN','',regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('nan','',regex=True)

df
```

```python
# Split Address data into street, state, and zip code columns
df[["Street", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)

df
```

```python
# Standardize Paying Customer and Do_Not_Contact columns
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df
```

```python
# Remove N/a and NaN values across full table
df = df.fillna('')
df = df.replace("N/a", "")

# Fill in blank values with default No for Do_Not_Contact column
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("", "N")
df
```

```python
# Remove rows for "Y" Do_Not_Call entries
for i in df.index:
    if df.loc[i,"Do_Not_Contact"] == "Y":
        df.drop(i, inplace=True)

# Remove rows for no Phone_Number entries
for i in df.index:
    if df.loc[i,"Phone_Number"] == "":
        df.drop(i, inplace=True)

df
```
